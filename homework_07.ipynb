## Le Monde

```python
from bs4 import BeautifulSoup
import requests

response = requests.get("https://www.lemonde.fr/en/")
doc = BeautifulSoup(response.text)
```

```python
# testing tags
doc.find('h1').get_text()
```

```python
# putting the parent item on a list
items = doc.find_all('div', attrs={'class':'lmd-link-clickarea'})
```

```python
# scraping the page and turning into another list
rows = []
for item in items:
    row = {}
    # Getting the titles
    row['title'] = item.find(class_='article__title').text.strip()
    # Getting the subheads
    try:
        row['subhead'] = item.find(class_='article__desc').text.strip()
    except:
        pass
    # Getting article URLs
    row['urls'] = item.find('a')['href']
    # Getting the premiums
    if item.find(class_='icon__premium icon--outside-simple'):
        row['premium'] = "That's a premium article."
    else:
        print('')
    # Getting the bylines
    try:
        row['bylines'] = item.find(class_='article__byline').text.strip()
    except:
        pass
    # Getting the image URL
    try:
        row['image_urls'] = item.find('img').get('data-src', None)
    except:
        pass
    rows.append(row)
```

```python
len(rows)
```

```python
# converting into a dataframe
import pandas as pd

df = pd.json_normalize(rows)
df
```

```python
# converting to a csv file
df.to_csv('articles.csv',index=False)
```

## US Courts

```python
from bs4 import BeautifulSoup
import requests

response3 = requests.get("https://www.tnwb.uscourts.gov/Search/Search.aspx?zoom_query=car&zoom_page=1&zoom_per_page=500&zoom_and=1&zoom_sort=0&zoom_xml=0")
doc3 = BeautifulSoup(response3.text)
```

```python
doc3
```

```python
# finding tags of interest
doc3.find_all('div', attrs= {'class':'results'})
```

```python
items = doc3.find_all(['div'], attrs= ['result_block', 'result_altblock'])
```

```python
# finding the tags we wanted
for item in items:
    print('-----')
    # Getting the urls
    try:
        print(item.find('a')['href'])
    except:
        pass
    # Getting the case
    try:
        print(item.find('a').text)
    except:
        pass
    # Getting the category
    print(item.find('span', attrs = {'class':'category'}).text)
    # Getting the context
    print(item.find('div', attrs = {'class':'context'}).text)
```

```python
# scraping

rows = []
for item in items:
    row = {}
    # Getting the urls
    try:
        row['urls'] = item.find('a')['href']
    except:
        pass
    # Getting the case
    try:
        row['case'] = item.find('a').text
    except:
        pass
    # Getting the category
    row['category'] = item.find('span', attrs = {'class':'category'}).text
    # Getting the additional details
    row['additional_details'] = item.find('div', attrs = {'class':'infoline'}).text
    rows.append(row)
```

```python
len(rows)
```

```python
import pandas as pd

df = pd.json_normalize(rows[1:])
df
```

```python
# spliting details into multiple columns
df[['matched', 'size', 'pdf_url']] = df['additional_details'].str.split('-', expand=True)
```

```python
df.head()
```

```python
df = df.drop('additional_details', axis = 1)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['pdf_url'] = df['pdf_url'].str.replace(' \xa0URL: ', '', regex=False)
```

```python
df.head()
```

```python
# converting pdf's columns into a list
pdf_list = df['pdf_url'].to_list()
```

```python
len(pdf_list)
```

```python
pdf_list
```

```python
import os
```

```python
# downloading the pdfs with the same name of the file
for url in pdf_list:
    response = requests.get(url)
    filename = os.path.basename(url)
    with open(filename, "wb") as file:
        file.write(response.content)
```
